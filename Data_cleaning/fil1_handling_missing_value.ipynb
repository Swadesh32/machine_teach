```python
 null_var = df.isnull().sum()/df.shape[0]*100    # for missing value parsentage
 null_var   
```

```python
drop_col = null_var[null_var > 11].keys()   # 11% se jada vale sari missing col
drop_col
```

```python
df2_drop_col = df.drop(columns = drop_col)
```

```python
df2_drop_col.shape
```

```python
plt.figure(figsize =(16,9))
sns.heatmap(df2_drop_col.isnull())
```

```python
df3_drop_rows = df2_drop_col.dropna() #dropna = ye us sari rowsko drop karta jisme nan ho
```

```python
num_var =['Year', 'Access to electricity (% of population)',
       'Access to clean fuels for cooking',
       'Renewable energy share in the total final energy consumption (%)',
       'Electricity from fossil fuels (TWh)', 'Electricity from nuclear (TWh)',
       'Electricity from renewables (TWh)',
       'Low-carbon electricity (% electricity)',
       'Primary energy consumption per capita (kWh/person)',
       'Energy intensity level of primary energy (MJ/$2017 PPP GDP)',
       'gdp_growth', 'gdp_per_capita', 'Land Area(Km2)', 'Latitude',
       'Longitude']

plt.figure(figsize =(16,9))
for i , var in enumerate(num_var):
    plt.subplot(5,3,i+1)
    sns.distplot(df[var],bins =20) 
    sns.distplot(df3_drop_rows[var],bins=20)
```

```python
df['Entity'].value_counts()/df3_drop_rows.shape[0]*100,df3_drop_rows['Entity'].value_counts()/df3_drop_rows.shape[0]*100
# theori ke form me distribution
```

```python
pd.concat([df['Entity'].value_counts()/df.shape[0]*100,
           df3_drop_rows['Entity'].value_counts()/df3_drop_rows.shape[0]*100],axis =1,
          keys =['Entity_org','Entity_clean'])
# key me new name h
```

```python
def cat_var_dist(var):
    return pd.concat([df[var].value_counts()/df.shape[0]*100,
              df3_drop_rows[var].value_counts()/df3_drop_rows.shape[0]*100],axis =1,
              keys =[var+'_org',var+'_clean'])

```
