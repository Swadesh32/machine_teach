# Ploting histograms

```python
import matplotlib.pyplot as plt 
import numpy as np
import random 
```

```python
ml_student_age = np.random .randint(18,45,(100))
py_student_age = np.random .randint(15,40,(100))
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.title('ML STUDENT AGE HISTOGRAME ')
plt.xlabel("Student of age ")
plt.ylabel("No of student ")
```

```python
plt.hist(
    x,
    bins=None,
    range=None,
    density=False,
    weights=None,
    cumulative=False,
    bottom=None,
    histtype='bar',
    align='mid',
    orientation='vertical',
    rwidth=None,
    log=False,
    color=None,
    label=None,
    stacked=False,
    *,
    data=None,
    **kwargs,
)
bins : int or sequence or str,
range : tuple or None, default
```

```python
bins =[15,20,25,30,35,40,45]
#plt.hist(ml_student_age, bins ,rwidth = 0.5 ,histtype ="step")
#plt.hist(ml_student_age, bins ,rwidth = 0.7 ,histtype ="bar")
plt.hist(ml_student_age, bins ,rwidth = 0.7 ,histtype ="bar",align ="right", orientation = "horizontal",    
         color = 'y',label= "ML student" )

plt.title('ML STUDENT AGE HISTOGRAME ')
plt.legend()
plt.xlabel("Student of age ")
plt.ylabel("No of student ")
```

```python
bins =[15,20,25,30,35,40,45]
plt.figure(figsize =(16,9))
#plt.hist(ml_student_age, bins ,rwidth = 0.5 ,histtype ="step")
#plt.hist(ml_student_age, bins ,rwidth = 0.7 ,histtype ="bar")
plt.hist([ml_student_age,py_student_age], bins ,rwidth = 0.7 ,histtype ="bar",align ="right", orientation = "vertical",    
         color = ['y',"m"],label= ["ML student","Py student"] )
#plt.hist(py_student_age, bins ,rwidth = 0.7 ,histtype ="bar",align ="right", orientation = "vertical",    
        # color = 'r',label= "ML student" )

plt.title('ML STUDENT AGE HISTOGRAME ')
plt.legend()  # label ke sath plt.legent ka use karte h 
plt.xlabel("Student of age ")
plt.ylabel("No of student ")
```
