# plotting Bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
cls =["python","AI","ML","DS",'R']  # must be no of item is same 
cls1_student = [40,50,30,20,10]  #outof 100 student in each class
cls2_student = [2,45,55,30,20]
cls3_student = [5,10,20,30,44]
```

```python
plt.barh(cls ,cls1_student)
```

```python
plt.bar(cls ,cls1_student)
```

```python
style.use("ggplot")
plt.figure(figsize =(5,5))
plt.bar(cls ,cls1_student ,width =0.2, align ="edge", color ="y", edgecolor = "m" ,linewidth = 5,alpha =0.5,
       linestyle ="--", label ='class1 student', visible = "Ture")
plt.legend()
```

```python
plt.bar(cls ,cls1_student ,width =0.2, align ="edge", color ="y", edgecolor = "m" ,linewidth = 5,alpha =0.5,
       linestyle ="--", label ='class1 student', visible = False)
plt.legend()
```

```python
plt.figure(figsize =(5,5))
plt.bar(cls ,cls1_student , align ="edge", color ="y", edgecolor = "m" ,linewidth = 4,alpha =0.5,
       linestyle ="-", label ='class1 student', visible = "Ture")
plt.bar(cls ,cls2_student , align ="edge", color ="y", edgecolor = "m" ,linewidth = 4,alpha =0.5,
       linestyle ="-", label ='class1 student', visible = "Ture")
plt.legend()
plt.title("Bar chart of iaip class",fontsize = 14)
plt.xlabel("classes",fontsize =14)
plt.ylabel("No of students",fontsize =15)
plt.show()
```

```python
plt.figure(figsize =(16,9))
cls_index= np.arange(len(cls))
plt.bar(cls_index,cls1_student,width ,color ="b",label ="class1 student ")
plt.bar(cls_index + width ,cls2_student,width ,color ="g",label ="class2 student ")
plt.bar(cls_index + width +width ,cls3_student,width ,color ="y",label ="class3 student ")

plt.xticks(cls_index + width, cls,rotation =20)
plt.title("Bar chart of iaip ",fontsize =18)
plt.xlabel("classes", fontsize =15)
plt.ylabel("No ofstudent",fontsize =15)
plt.show()
```

```python
plt.figure(figsize =(16,9))
cls_index= np.arange(len(cls))
plt.barh(cls_index,cls1_student,width ,color ="b",label ="class1 student ")
plt.barh(cls_index + width ,cls2_student,width ,color ="g",label ="class2 student ")
plt.barh(cls_index + width +width ,cls3_student,width ,color ="y",label ="class3 student ")

plt.yticks(cls_index + width, cls,rotation =20)
plt.title("Bar chart of iaip ",fontsize =18)
plt.ylabel("classes", fontsize =15)
plt.xlabel("No ofstudent",fontsize =15)
plt.legend()
plt.show()
```
