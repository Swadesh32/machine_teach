```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
dt = pd.read_csv("C:\\Users\\maury\\Desktop\\csv\\dataset1.csv",nrows =30)         # tole no of rows is 32 
dt.shape
```

```python
dt
```

```python
x = dt["Class"]
y = dt["JNV"]
z = dt["KV"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y,color ="r", )
#plt.scatter(x,z ,color = "g")
plt.title('STUDENT DATA FOR ACRRODING TO INFORMATION',fontsize = 15)
plt.xlabel('Classes ofstudent',fontsize = 13)
plt.ylabel("JNV of student",fontsize =13)
```

```python
plt.scatter(
    x,
    y,
    s=None,
    c=None,
    marker=None,
    cmap=None,
    norm=None,
    vmin=None,
    vmax=None,
    alpha=None,
    linewidths=None,
    *,
    edgecolors=None,
    plotnonfinite=False,
    data=None,
    **kwargs,
)
```

```python
plt.figure(figsize= (3,3))
#plt.scatter(x,y,c ="r", marker ="*" ,s =15,alpha =0.5,linewidths =5,edgecolor ="g")
plt.scatter(x,y,c ="r" ,s =15,alpha =0.5,linewidths =5,edgecolor ="g")

plt.title('STUDENT DATA FOR ACRRODING TO INFORMATION',fontsize = 15)
plt.xlabel('Classes ofstudent',fontsize = 13)
plt.ylabel("JNV of student",fontsize =13)
```

```python
# tow scater in plot

plt.figure(figsize= (3,3))
plt.scatter(x,y,c ="r", marker ="*" ,s =15,alpha =0.5,linewidths =5,edgecolor ="g")
plt.scatter(x,z,c ="y", marker ="*" ,s =15,alpha =0.5,linewidths =5,edgecolor ="y")




plt.title('STUDENT DATA FOR ACRRODING TO INFORMATION',fontsize = 15)
plt.xlabel('Classes ofstudent',fontsize = 13)
plt.ylabel("KV and JNV of student",fontsize =13)
```
