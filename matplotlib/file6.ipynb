```python
plt.figure(figsize =(16,9))

#plt.subplot(2,3,1)
#________________________________________________________________________________________________________
plt.subplot(231)

import matplotlib.pyplot as plt 
import numpy as np
import random 
ml_student_age = np.random .randint(18,45,(100))
py_student_age = np.random .randint(15,40,(100))
plt.hist(ml_student_age)
plt.title('ML STUDENT AGE HISTOGRAME ')
plt.xlabel("Student of age ")
plt.ylabel("No of student ")

#________________________________________________________________________________________________________
plt.subplot(2,3,2)
plt.subplot(2,3,3)
plt.subplot(2,3,5,projection ="polar", facecolor ="k" , frameon =True)
plt.show()
```
