# fillna mathods 

```python
#pandas fillna --> fillna fill the NAN values with given values (input)
```

```python
import pandas as  pd 
```

```python
df2 = pd.read_csv('C:\\Users\\maury\\Desktop\\csv\\Book5.csv')
df2
```

```python
df2.fillna(0)
```

```python
df2.fillna('a')
```

```python
df2.fillna({'st_id.2': 'none','data': 'empty'})
```

```python
df2.fillna(method = "ffill") # privious(frist fill) values ko karta h 
```

```python
df2.fillna(method = "bfill",axis =1) 
```

```python
df2.fillna(0, limit = 1) 
```
